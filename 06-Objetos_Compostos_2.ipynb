## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import pandas as pd
import numpy as np

dados_de_seguro_na_florida_em_2012 = pd.read_csv('data.csv')
dados_de_seguro_na_florida_em_2012.head()
```

```python


dicionario = {
    "numero_de_entradas" : dados_de_seguro_na_florida_em_2012.count() ,
    "classe_das_colunas" : dados_de_seguro_na_florida_em_2012.iloc[0].apply(type),
    "numero_de_elementos_de_cada_classe" : dados_de_seguro_na_florida_em_2012.replace(0, np.nan).nunique(),
    "media_por_coluna" : dados_de_seguro_na_florida_em_2012.select_dtypes(include="number").iloc[:, [0, 3, 8]].mean(),
    "desvio_padrao_por_coluna" : dados_de_seguro_na_florida_em_2012.select_dtypes(include="number").iloc[:, [0, 3, 8]].std(),
    "média_por_coluna" : dados_de_seguro_na_florida_em_2012.iloc[:, [0, 3, 8]].quantile(0.75)
}

dicionario
```
