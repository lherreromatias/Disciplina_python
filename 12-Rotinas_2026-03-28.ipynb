Quantos e quais são os valores superiores a 10?

```python
import numpy as np
import pandas as pd

def valor_maior_que (array_valor: np.ndarray, i: int) -> list:
    array_rand_filtro = []
    for valor in array_valor:
        if valor > i:
            array_rand_filtro.append(float(valor))
    return array_rand_filtro

def qnt_maior_que (array_valor: np.ndarray, i: int) -> int:
    a = 0
    for valor in array_valor:
        if valor > i:
            a += 1
    return a

array_rand = np.random.normal(10,3,5)
lisa_maior_que = valor_maior_que(array_rand, 10)
numero_maior_que = qnt_maior_que(array_rand, 10)


print("maiores que - ",lisa_maior_que, "\n","numero de valores superiores a - ", numero_maior_que)
array_rand
```

```python
lista_numeros = np.random.randint(0, 10, size=5).tolist()
lista_numeros

def media (lista_numeros: list) -> float:
    n = len(lista_numeros)
    s = sum(lista_numeros)
    try:
        m = s/n
        return m
    except ZeroDivisionError:
        return "Erro: da para dividir por zero não"
    except ZeroDivisionError:
        return "Erro: os objetos compativeis para a divisão"
    
print(f"os valores são {lista_numeros}, a media é deles é {media(lista_numeros)}")
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
obj1 = 1
obj2 = 2

def diviao_obj_simpl (obj1: int, obj2: int) -> float:
    print(f"Objeto 1: Classe {type(obj1)} | Valor: {obj1}")
    print(f"Objeto 2: Classe {type(obj2)} | Valor: {obj2}")
    try:
        d = obj1/obj2
        return d
    except ZeroDivisionError:
        return "Erro: da para dividir por zero não"
    except ZeroDivisionError:
        return "Erro: os objetos compativeis para a divisão"

print(f"\nos objetos são {obj1} e {obj2}, a razão deles é {diviao_obj_simpl(obj1,obj2)}")


```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
obj1 = np.random.randint(0, 10, size=5).tolist()
obj2 = np.random.randint(1, 10, size=5)

def diviao_obj_simpl (obj1: int, obj2: int) -> float:
    print(f"Objeto 1: Classe {type(obj1)} | Valor: {obj1}")
    print(f"Objeto 2: Classe {type(obj2)} | Valor: {obj2}")

    val_obj1 = obj1.values if hasattr(obj1, 'values') else obj1
    val_obj2 = obj2.values if hasattr(obj2, 'values') else obj2

    if 0 in val_obj2:
        return "Erro: Existe um zero no divisor."

    try:
        d = (val_obj1/val_obj2).tolist()
        return d
    except ZeroDivisionError:
        return "Erro: da para dividir por zero não"
    except ZeroDivisionError:
        return "Erro: os objetos compativeis para a divisão"

print(f"\nos objetos são {obj1} e {obj2.tolist()}, a razão deles é {diviao_obj_simpl(obj1,obj2)}")
```
