## 03 – Usando o método DataFrame.to_csv(‘nome_do_arquivo.csv’) exporte o objeto obtido no exercício 10 da aula anterior para um arquivo CSV. Usando o método pd.read_csv(‘nome_do_arquivo.csv’) instancie uma DataFrame. Use o valor ‘Unnamed: 0’ no parâmetro index_col, para ajustar o index. Arredonde os valores para até três casas decimais, usando o método round().
Recupere a DataFrame.

```python
Ultima_coluna = pd.read_csv('ultima_Coluna.csv', index_col='Unnamed: 0')
Ultima_linha = pd.read_csv('ultima_linha.csv', index_col='Unnamed: 0')
Ultima_coluna = Ultima_coluna.round(3)
Ultima_linha = Ultima_linha.round(3)
print(Ultima_coluna,"\n\n",Ultima_linha)
```

## 04 – Instancie um numpy array que apresente o mesmo formato (linhas e colunas) da DataFrame. Este array deve conter apenas inteiros uns.
Recupere o array.

Obs.: Procure na internet o método numpy.ones()

```python
tam_linha = Ultima_linha.shape
tam_coluna = Ultima_coluna.shape
array_linha_uns = np.ones(tam_linha, dtype=int)
array_coluna_uns = np.ones(tam_coluna, dtype=int)
print(" tamanho colunas\n",array_coluna_uns,"\n\n","tamaho linhas \n",array_linha_uns)
```

## 05 – Quantos dos valores contidos na DataFrame são iguais aos valores no array?

```python
comp_linha = array_linha_uns == Ultima_linha.values
comp_linha.sum()
comp_coluna = array_coluna_uns == Ultima_coluna.values
comp_coluna.sum()
print(" array linhas \n",comp_linha.sum(),"\n\n","array colunas\n",comp_coluna.sum())
```

## 06 – Quantos dos valores contidos na DataFrame são superiores aos valores no array?

```python
comp_linha = array_linha_uns < Ultima_linha.values
comp_linha.sum()
comp_coluna = array_coluna_uns < Ultima_coluna.values
comp_coluna.sum()
print(" array linhas \n",comp_linha.sum(),"\n\n","array colunas \n",comp_coluna.sum())
```

## 07 – Quantos dos valores contidos na DataFrame são iguais ou inferiores aos valores no array?

```python
comp_linha = array_linha_uns >= Ultima_linha.values
comp_linha.sum()
comp_coluna = array_coluna_uns >= Ultima_coluna.values
comp_coluna.sum()
print(" array linhas \n",comp_linha.sum(),"\n\n","array colunas \n",comp_coluna.sum())
```

## 08 – Quantos dos valores contidos na DataFrame estão entre 0 e 1?

```python
comp_linha = (0 < Ultima_linha.values) & (Ultima_linha.values < 1)
comp_linha.sum()
comp_coluna = (0 < Ultima_coluna.values) & (Ultima_coluna.values < 1)
comp_coluna.sum()
print(" array linhas \n",comp_linha.sum(),"\n\n","array colunas \n",comp_coluna.sum())
```
