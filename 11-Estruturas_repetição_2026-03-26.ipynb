## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Primeiramente foram definidas 3 listas vazias, uma para conter os 5000 valores, uma para conter somente os valores pares e outra para conter os valores ímpares. Ainda foi instanciado um objeto contador com valor zero 
Então, foi instanciado um “while” que ficaria em looping até o valor do contador deixar de ser diferente de 5000. Em cada loop, um valor aleatório, entre 1 e 5000, era atribuído a uma variável “a” e essa variável era adicionada à lista total. Então era conferido se a variável “a” desse loop era par ou ímpar, e então ela era adicionada a respectiva lista. Ainda em cada loop era somado 1 ao valor do contador. 
Quando o contador chega em 5000, o while cessa, e então é impresso o tamanho de cada lista.

## 03 – Escreva o programa em Python:

```python
import numpy as np
import pandas as pd
import random

contador = 0
lista5000 = []
lista5000_par = []
lista5000_impar = []

while contador != 5000 :
    a = random.randint(0, 5000)
    lista5000.append(a)
    if a % 2 == 0 :
        lista5000_par.append(a)
    if a % 2 != 0 :
        lista5000_impar.append(a)

    contador = contador + 1
    #print(contador)

print(" tamanho da lista - ",len(lista5000),"\n","numero de objetos pares - ", len(lista5000_par),"\n","numero de objetos impares - ",len(lista5000_impar))
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
gene_dataframe = pd.read_csv("gene_dataframe.csv", index_col='Unnamed: 0')
print(gene_dataframe, "\n\n\n")
a = 0
c = 0

while c != 5:
        numb = gene_dataframe.iloc[c,0]
        #print (numb)
        if (numb % 2 == 0) :
            print("é par - ",gene_dataframe.index[c])

        if (numb % 2 != 0) :
            a += 1
        
        c += 1

print("Ha",a,"genes com numero de letas impares")
```
