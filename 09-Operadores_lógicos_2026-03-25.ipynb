## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
import numpy as np
import pandas as pd
import math as mt

p = 2.0 + (3.0/5.0)
q = (p + 10.2) - (5.0 * 2.0)
r = (p/q) + 33.8
s = p/(q + r)
t = s * 2

print(" p == 10", p == 10, "\n",
    "q <= 2.9", q <= 2.9, "\n",
    "10 <= r < 100", 10 <= r and r < 100, "\n",
    "s != 5", s != 5, "\n",
    "t == 33", t == 33, "\n",
)

print("para inverter os operadores lógicos podemos - \n"
    " p != 10", p != 10, "\n",
    "q >= 2.9", q >= 2.9, "\n",
    "10 >= r > 100", 10 >= r and r > 100, "\n",
    "s == 5", s == 5, "\n",
    "t != 33", t != 33, "\n",
)
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]


print("o resultado da comparação p or q or r or s or t or u é de -",p or q or r or s or t or u)

print("para inverter o resultado com o not podemos usar  not (p or q or r or s or t or u) -", not (p or q or r or s or t or u))
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

```python
Ultima_coluna = pd.read_csv('ultima_Coluna.csv', index_col='Unnamed: 0')
Ultima_linha = pd.read_csv('ultima_linha.csv', index_col='Unnamed: 0')
Ultima_coluna = Ultima_coluna.round(3)
Ultima_linha = Ultima_linha.round(3)
print("ultima_coluna -\n", Ultima_coluna, "\n\n\n", "ultima_linha -\n", Ultima_linha,"\n\n")
print("ultima_coluna - [1,:] == [2,:] ou [2,:] == [3,:] -\n ", (Ultima_coluna.iloc[1,:] == Ultima_coluna.iloc[2,:]).all() or (Ultima_coluna.iloc[2,:] == Ultima_coluna.iloc[3,:]).all())
print("ultima_linha - [1,:] == [2,:] ou [2,:] == [3,:] -\n ", (Ultima_linha.iloc[1,:] == Ultima_linha.iloc[2,:]).all() or (Ultima_linha.iloc[2,:] == Ultima_linha.iloc[3,:]).all())
```

## 04 – Usando operadores verifique se todos os valores contidos na primeira linha são maiores ou iguais a 10.

```python
print("ultima_coluna - [0,:] >= 10 -\n ", (Ultima_coluna.iloc[0,:] >= 10).all())
print("ultima_linha - [0,:] >= 10 -\n ", (Ultima_linha.iloc[0,:] >= 10).all())
```

## 05 – Usando operadores verifique se todos os valores contidos na terceira coluna estão entre 0 inclusive e 40.

```python
print("ultima_coluna - 0 < [:,3] < 40 -\n ", (Ultima_coluna.iloc[:,3] > 0).all() and (Ultima_coluna.iloc[:,3] < 40).all())
print("ultima_linha - 0 < [:,3] < 40 -\n ", (Ultima_linha.iloc[:,3] > 0).all() and (Ultima_linha.iloc[:,3] < 40).all())
```
