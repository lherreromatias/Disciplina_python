# Exercício 07:

Identifique a classe de objeto em Python dos seguintes valores:

(a) "9 de agosto de 1968"
<br>R.: String
<br>
<br>
(b) 1.3
<br>R.: Float
<br>
<br>
(c) False
<br>R.: Bool
<br>
<br>
(d) -31
<br>R.: Int
<br>
<br>
(e) “?”
<br>R.: String
<br>
<br>
(f) ‘1979’
<br>R.: string
<br>
<br>

```python
dict  = {
    "a" : "9 de agosto de 1968",
    "b" : 1.3,
    "c" : False,
    "d" : -31,
    "e" : "?",
    "f" : '1979'}
for chave, valor in dict.items():
    print(chave,":", type(valor))
```
