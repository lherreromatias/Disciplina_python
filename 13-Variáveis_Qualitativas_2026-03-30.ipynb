```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

gene_dataframe = pd.read_csv("gene_dataframe.csv", index_col='Unnamed: 0')
print(gene_dataframe, "\n\n\n")

tabela_freq = gene_dataframe['num_letras'].value_counts()
gene_dataframe['frequencia_do_tamanho'] = gene_dataframe['num_letras'].map(tabela_freq)


tabela_freq.sort_index().plot(kind='bar', color='orchid', edgecolor='black')
import matplotlib.pyplot as plt
plt.title('Frequência do Número de letras')
plt.xlabel('Número de letras')
plt.ylabel('número de Genes')
plt.yticks([0, 1, 2])
plt.show()
```

## 07 – Reorganize a DataFrame, removendo as colunas contendo os dois primeiros caracteres, o número de caracter 'a', e a presença de 'hu', dos genes. Renomeie as colunas de acordo com as informações. Adicione uma coluna com a informação de 'Amostra 1'.

```python
remove = ['primeiras_duas_letras', 'num_a', 'contains_hu']
gene_dataframe = gene_dataframe.drop(columns=remove)

novos_nomes = {
    'num_letras': 'Comprimento do Nome',
    'frequencia_do_tamanho': 'Frequência do Comprimento'
}
gene_dataframe = gene_dataframe.rename(columns=novos_nomes)
gene_dataframe['Amostra_1'] = 'Amostra 1'
display(gene_dataframe)
```

## 08 – Usando a função randint(), da classe random do módulo numpy (np.random.randint(1,10,5)), instancie objeto da classe numpy array. Calcule as frequências desta nova amostra, considerando que à uma ordem entre os genes. Insira na DataFrame este objeto como "Amostra 2".

```python
amostra_1 = np.random.randint(1, 10, 5)
tabela_freq = pd.Series(amostra_1).value_counts()
gene_dataframe['Amostra_1'] = pd.Series(amostra_1).map(tabela_freq).values
display(gene_dataframe)

amostra_2 = np.random.randint(1, 10, 5)
tabela_freq = pd.Series(amostra_2).value_counts()
gene_dataframe['Amostra_2'] = pd.Series(amostra_2).map(tabela_freq).values
display(gene_dataframe)
```

## 09 – Repita o exercício 8, inserindo na DataFrame as replicas "Amostra 3“, "Amostra 4“, "Amostra 5“ e "Amostra 6".

```python
a = 2
while a != 7:
    amostra = np.random.randint(1, 10, 5)
    tabela_freq = pd.Series(amostra).value_counts()
    gene_dataframe[f'Amostra_{a}'] = pd.Series(amostra).map(tabela_freq).values
    a += 1
display(gene_dataframe)
```

## 10 – Obtenha dos seus dados o gráfico abaixo, adicionando 3 replicas.

```python
while a != 10:
    amostra = np.random.randint(1, 10, 5)
    tabela_freq = pd.Series(amostra).value_counts()
    gene_dataframe[f'Amostra_{a}'] = pd.Series(amostra).map(tabela_freq).values
    a += 1
display(gene_dataframe)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_longo = gene_dataframe.reset_index().melt(
    id_vars=['index'], 
    value_vars=['Amostra_1', 'Amostra_2', 'Amostra_3', 'Amostra_4', 
                'Amostra_5', 'Amostra_6', 'Amostra_7', 'Amostra_8', 'Amostra_9'],
    var_name='Amostra_Original', 
    value_name='Frequencia'
)

df_longo['Frequencia'] = pd.to_numeric(df_longo['Frequencia'], errors='coerce')
df_longo['index'] = df_longo['index'].astype(str)

mapeamento = {
    'Amostra_1': 'Replica 1', 'Amostra_2': 'Replica 1', 'Amostra_3': 'Replica 1',
    'Amostra_4': 'Replica 2', 'Amostra_5': 'Replica 2', 'Amostra_6': 'Replica 2',
    'Amostra_7': 'Replica 3', 'Amostra_8': 'Replica 3', 'Amostra_9': 'Replica 3'
}
df_longo['amostras'] = df_longo['Amostra_Original'].map(mapeamento)

plt.figure(figsize=(12, 7))
sns.set_style("white")

sns.barplot(
    data=df_longo, 
    x='amostras', 
    y='Frequencia', 
    hue='index', 
    palette='muted',
    capsize=.05,
    errorbar='sd'
)

plt.title('Frequências Absoluta Simples por Replica e Réplica')
plt.ylabel('Frequência Absoluta Simples')
plt.xlabel('amostras')
plt.legend(title='Genes', bbox_to_anchor=(1, 0.5), loc='center left')
sns.despine()

plt.show()
```
