## Instancie um objeto da classe array numpy, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3. 
Quantos e quais são os valores superiores a 10?

```python
array_rand = np.random.normal(10,3,5)
array_rand_filtro = []

for valor in array_rand:
    if valor > 10:
        array_rand_filtro.append(array_rand_filtro)
        array_rand_filtro = [valor]

array_rand_filtro = np.array(array_rand_filtro)
print("array - ",array_rand, "\n","somente maiores de 10 - ", array_rand_filtro)

        
```
